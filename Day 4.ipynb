#### Sort by multiple columns (Department, then Salary).

```python
import pandas as pd

# Sample DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Department": ["HR", "IT", "IT", "HR", "Finance"],
    "Salary": [50000, 70000, 60000, 55000, 65000]
}

df = pd.DataFrame(data)

# Sort first by Department, then by Salary
df_sorted = df.sort_values(by=["Department", "Salary"])
print(df_sorted)
```

```python
df_sorted = df.sort_values(by=["Department", "Salary"], ascending=[True, False])
print(df_sorted)
```
