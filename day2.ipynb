#### Filter rows where Salary is between 50,000 and 80,000.

```python
names = [f"Person{i}" for i in range(1, 51)]

# Generate 50 random salaries between 20000 and 120000
salaries = np.random.randint(20000, 120001, size=50)

# Create DataFrame
df = pd.DataFrame({
    "Name": names,
    "Salary": salaries
})
pd.set_option("display.max_rows", None)
print(df)

```

```python
(df[(df["Salary"]>50000)&(df["Salary"]<100000)])[["Name","Salary"]]
```

```python
filtered = df[(df["Salary"] >= 50000) & (df["Salary"] <= 80000)]
print(filtered.to_string())
```
